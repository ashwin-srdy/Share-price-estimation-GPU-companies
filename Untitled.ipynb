```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt

for i in amd.columns[1:]:
    print(i)
    plt.rcParams.update({'figure.figsize':(9,7), 'figure.dpi':120})

    # Original Series
    fig, axes = plt.subplots(3, 2, sharex=True)
    axes[0, 0].plot(amd[i]); axes[0, 0].set_title('Original Series')
    plot_acf(amd[i], ax=axes[0, 1])

    # 1st Differencing
    axes[1, 0].plot(amd[i].diff()); axes[1, 0].set_title('1st Order Differencing')
    plot_acf(amd[i].diff().dropna(), ax=axes[1, 1])

    # 2nd Differencing
    axes[2, 0].plot(amd[i].diff().diff()); axes[2, 0].set_title('2nd Order Differencing')
    plot_acf(amd[i].diff().diff().dropna(), ax=axes[2, 1])

    plt.show()
    print()
```

```python
# Split data into training and testing sets
train_size = int(len(amd) * 0.8)
train, test = amd['Low'].iloc[:train_size], amd['Low'].iloc[train_size:]
print(train.shape)
print(test.shape)
```

```python
# Fit ARIMA model
model = ARIMA(train, order=(2, 2, 1))
model_fit = model.fit()

# Make predictions on testing set
predictions = model_fit.forecast(len(test))
```

```python
# RMSE
mae = mean_absolute_error(test, predictions)
r2 = r2_score(test, predictions)
print('MAE: %.2f' % mae)
print('R2_score:',r2)

# Plot predictions and actual values
plt.figure(figsize=(15,8))
plt.plot(train)
plt.plot(test)
plt.plot(predictions, color='red')
plt.show()
```

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from sklearn.metrics import mean_squared_error

df=new

# Convert the index to datetime format
df.index = pd.to_datetime(df.index)

# Split the data into training and testing sets
train_data = df[:'2020']
test_data = df['2020':]

# Plot the training and testing data
plt.plot(train_data, label='Training Data')
plt.plot(test_data, label='Testing Data')
plt.legend()
plt.show()

# Plot the autocorrelation and partial autocorrelation plots
#plot_acf(train_data, lags=40)
#plot_pacf(train_data, lags=40)
#plt.show()

# Fit the SARIMA model to the training data
order=(1, 1, 0)
seasonal_order=(1, 1, 0, 12)
sarima_model = SARIMAX(train_data, order=order, seasonal_order=seasonal_order)
sarima_fit = sarima_model.fit()

# Make predictions on the testing data
sarima_predictions = sarima_fit.predict(start='2020-01-01', end='2020-12-31')

# Plot the actual vs predicted values
plt.plot(test_data, label='Actual Data')
plt.plot(sarima_predictions, label='SARIMA Predictions')
plt.legend()
plt.show()

# Calculate the root mean squared error (RMSE)
sarima_rmse = np.sqrt(mean_squared_error(test_data, sarima_predictions))
print('SARIMA RMSE:', sarima_rmse)
```
